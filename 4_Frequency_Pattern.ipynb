# Frequency Pattern
### This part I will show you which combination of heroes has high win rate or low win rate

```python
# Run some setup code for this notebook.
from __future__ import print_function

from efficient_apriori import apriori
import pandas as pd
import numpy as np
import json
```

```python
df=pd.read_csv("out.csv")
df.head(5)
```

```python
wintransactions=[]
losetransactions=[]
```

```python
from progressbar import ProgressBar
pbar = ProgressBar()
for i in pbar(range(0,len(df))):
    if(df.iloc[i,0]==True):
        wintransactions.append(list(df.iloc[i,23:28]))
        losetransactions.append(list(df.iloc[i,28:33]))
        
    if(df.iloc[i,0]==False):
        wintransactions.append(list(df.iloc[i,28:33]))
        losetransactions.append(list(df.iloc[i,23:28]))
```

```python
#Use apriori find the frequency parttern, then calculate the win rate of these patterns
itemsets, rules = apriori(wintransactions, min_support=0.005,  min_confidence=0)
itemsets1, rules1 = apriori(losetransactions, min_support=0.005,  min_confidence=0)
```

```python
itemsets1[2]
```

```python
#The name of heroes
heroesname=[]
for i in range(0,130):
    heroesname.append("empty")
    
fopen=open('heroes.json','r')
heroes=json.load(fopen)
fopen.close()

for i in range(0,len(heroes)):
    heroesname[heroes[i].get('id')]=heroes[i].get('localized_name')
```

```python
keys=[]
for key in itemsets[2]:
    if(itemsets1[2].get(key)!=None):
        keys.append(key)
```

```python
for i in range(0,len(keys)):
    rate=itemsets[2].get(keys[i])/(itemsets1[2].get(keys[i])+itemsets[2].get(keys[i]))*100
    if(rate>55):
        s=keys[i]
        print("Win rate of "+heroesname[s[0]]+ " and "+ heroesname[s[1]]+ " is\t " + str(rate) + "%")
```

```python
for i in range(0,len(keys)):
    rate=itemsets[2].get(keys[i])/(itemsets1[2].get(keys[i])+itemsets[2].get(keys[i]))*100
    if(rate<45):
        s=keys[i]
        print("Win rate of "+heroesname[s[0]]+ "\t and \t"+ heroesname[s[1]]+ "\t is\t " + str(rate) + "%")
```
