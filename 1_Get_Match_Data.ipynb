# Get Match Data

### This this the first part which is used to get the DOTA data. Basically, in this part you will use requests, steam API and match ID to get the detail of the match in JSON format.

```python
# Run some setup code for this notebook.
from __future__ import print_function

import requests
import json
from progressbar import ProgressBar
import time
```

```python
#Start a request session
c=requests.session()

#Use a List store the MatchID
MatchID=[]
```

### Here is the major part to use web crawler get the data. Here are some information you should notice
- **Key** is the steam dev API, you can use mine,but it's better to apply your own key
- The range(4479300000,4479300100) is the range of Match ID. Here is an example. To get more data, the range should be large. In my project, I spend more than 24 hours to get 120000 matches.
- Please make sure you have a data directory

```python
#pbar is helpful to show the progress
pbar = ProgressBar()
for i in pbar(range(4479300000,4479300100)):
    url='https://api.steampowered.com/IDOTA2Match_570/GetMatchDetails/V1/?key=D607273A4EF0E7CC28426E674FDFF761&match_id='+\
        str(i)
    l=c.get(url)
    a=l.json()
    if(len(a)==0):
        continue
    if(len(a['result'])>10):
        with open("./data/"+str(i)+".json","w") as f:
            json.dump(a,f)
        MatchID.append(i)
```
